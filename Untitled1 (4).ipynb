# Разделяем на Х и на Y где Х данные, а Y классы болезней

```python
y = pd.DataFrame(df['Class'])
x = pd.DataFrame(df.drop('Class', axis = 1))
```

from missingpy import MissForest
Намного лучше from sklearn.impute import SimpleImputer, но работает дольше.

```python
from sklearn.preprocessing import Imputer
#from sklearn.impute import SimpleImputer
import time
```

```python
start_time = time.time()

imp = Imputer(missing_values = np.nan, strategy = 'mean')
x_imp = imp.fit_transform(x)
x_imp = pd.DataFrame(x_imp, columns = x.columns)

end_time = time.time()
```

```python
end_time - start_time
```

```python
x_imp.head()
```

```python
max(x_imp.isnull().sum())
```

# Поиск информативных признаков

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
```

```python
start_time = time.time()

x_sel = SelectKBest(mutual_info_classif, k = 300).fit_transform(x_scaled, y)
x_sel = pd.DataFrame(x_sel)

end_time = time.time()
```

```python
end_time - start_time
```

```python
x_sel.head()
```

```python
x_end = x_sel
```

# Разбиваем на обучающую и тестовую выборку.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_end, y, test_size = 0.20, random_state = 42)
```

```python
type(y_train)
```

# Поиск оптимальных параметров

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.metrics import cohen_kappa_score
```

```python
tuned_parameters = [{'n_estimators': [100], 
                     'criterion': ['gini', 'entropy'],
                     'max_depth': [5, 6, 7],
                     'max_features': ['sqrt', 'log2', None]}]
```

```python
score = 'accuracy'
```

```python
start_time = time.time()
print("# Tuning hyper-parameters for %s" % score)
print()
clf = GridSearchCV(RandomForestClassifier(), tuned_parameters, cv = 5,
                   scoring = score)
clf.fit(x_train, y_train)

end_time = time.time()
```

```python
end_time - start_time
```

# для балансед акураси

```python
print("Best parameters set found on development set:")
print()
print(clf.best_params_)
```

# для акураси

```python
print("Best parameters set found on development set:")
print()
print(clf.best_params_)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
model = RandomForestClassifier(criterion = 'entropy', max_depth = 7, max_features = None, n_estimators = 100)

model.fit(x_train, y_train)
pred_test = model.predict(x_test)
```
